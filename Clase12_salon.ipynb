```python
def borracho(pasos):
    pasosDer = 0
    pasosIzq = 0
    x0 = 0
    posiciones = [x0]
    for paso in range(pasos):
        decision = np.random.rand()
        if decision < .5 :
            pasosIzq += 1 
            x0 -= 1
            posiciones.append(x0)
        else:
            pasosDer += 1
            x0 += 1
            posiciones.append(x0)
    return posiciones
```

```python
plt.figure(figsize=(13,10))
plt.xlabel("Pasos")
plt.ylabel("Pos")
for i in range(5):
    plt.plot(borracho(1000), lw=.9,label = "Boracho %s" %(i+1))
plt.legend()
plt.show;
```

```python
def yinf(x,y,n,xini,xfin,yini,yfin):
    j=0
    count=0
    for i in range(n):
        if (y[i]>yini>y[j]) or (y[i]<yini<y[j]):
                if xini <x[j]<xfin:
                    count += 1
        j = i
    return count

def ysup(x,y,n,xini,xfin,yini,yfin):
    j=0
    count=0
    for i in range(n):
        if (y[i]>yfin>y[j]) or (y[i]<yfin<y[j]):
                if xini <x[j]<xfin:
                    count += 1
        j = i
    return count

def xder(x,y,n,xini,xfin,yini,yfin):
    j=0
    count=0
    for i in range(n):
        if (x[i]>xfin>x[j]) or (x[i]<xfin<x[j]):
                if yini <y[j]<yfin:
                    count += 1
        j = i
    return count

def xizq(x,y,n,xini,xfin,yini,yfin):
    j=0
    count=0
    for i in range(n):
        if (x[i]>xini>x[j]) or (x[i]<xini<x[j]):
                if yini <y[j]<yfin:
                    count += 1
        j = i
    return count
```
